```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
        
```

```python
df1['genres']=df1['genres'].apply(convert)
```

```python
df1['keywords']=df1['keywords'].apply(convert)
```

```python
df1['keywords'].head()
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        
        if  counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
df1['cast']=df1['cast'].apply(convert3)
```

```python
df1['cast'].head(2)
```

```python
df1['crew'].head()
```

```python
def fetchdirector(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
df1['crew']=df1['crew'].apply(fetchdirector)
```

```python
df1['cast']=df1['cast'].apply(lambda x: [i.replace(" ",'') for i in x])
df1['crew']=df1['crew'].apply(lambda x: [i.replace(" ",'') for i in x])
df1['genres']=df1['genres'].apply(lambda x: [i.replace(" ",'') for i in x])
df1['keywords']=df1['keywords'].apply(lambda x: [i.replace(" ",'') for i in x])
```

#### STEMING

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
print(ps.stem('danced'))
print(ps.stem('dancing'))
print(ps.stem('dance'))
```

```python
def stem(text):
    l1=text.split()
    l2=[]
    for i in l1:
        l2.append(ps.stem(i))
    return ' '.join(l2)
        
    
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
df.tags.head(1)
```

#### COUNTVECTORIZER ON NEW DATAFRAME AFTER STEMMING

```python
from sklearn.feature_extraction.text import CountVectorizer
cv1=CountVectorizer(max_features=5000,stop_words='english')
vectorizer_new_df=cv1.fit_transform(df['tags']).toarray()
cv1.get_feature_names()
```

### COSINE SIMILARITY CAN BE USED HERE IN MEASURING SIMILARITY BETWEEN VECTORS INSTEAD OF EUCLEADIAN DISTANCE SINCE IT IS DOES NOT WORK WITH HIGH DIMENSIONALITY 

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectorizer_new_df)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index=df[df['title']==movie].index[0]
    movies_list=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(df['title'].iloc[i[0]])
        
    
```

```python
recommend('Batman Begins')
```
